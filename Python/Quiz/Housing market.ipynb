### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Create 3 differents Series, each of length 100, as follows: 
1. The first a random number from 1 to 4 
2. The second a random number from 1 to 3
3. The third a random number from 10,000 to 30,000

```python
random1 = np.random.randint(1,4,100)
random1
```

```python
random2 = np.random.randint(1,3,size=100)
random2
```

```python
random3 = np.random.randint(10000,30000,100)
random3
```

### Step 3. Let's create a DataFrame by joinning the Series by column

```python
rand1 = pd.DataFrame(random1,columns=['random_number1'])
rand1.head(5)
```

```python
rand2 = pd.DataFrame(random2,columns=['random_number2'])
rand2.head(5)
```

```python
rand3 = pd.DataFrame(random3,columns=['random_number3'])
rand3.head(5)
```

```python
all_rand = pd.concat([rand1,rand2,rand3],axis=1)
all_rand.head(5)
```

### Step 4. Change the name of the columns to bedrs, bathrs, price_sqr_meter

```python
all_rand.columns=['bedrs','bathrs','price_sqr_meter']
all_rand.columns
```
